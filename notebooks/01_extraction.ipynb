# 01 — Data Extraction

**Objective:** Load the raw dataset, inspect its structure, and confirm it is suitable for analysis.

**Dataset Source:** [Online Sales Dataset — Kaggle](https://www.kaggle.com/datasets/yusufdelikkaya/online-sales-dataset)

**Problem Statement:** Analyse product returns in e-commerce to identify patterns, quantify revenue leakage, and generate actionable business recommendations.

```python
import pandas as pd
import numpy as np
import os
```

## 1. Load Raw Dataset

```python
RAW_PATH = '../data/raw/online_sales_dataset.csv'

# Verify file exists
assert os.path.exists(RAW_PATH), f'File not found: {RAW_PATH}'
print(f'File size: {os.path.getsize(RAW_PATH) / 1e6:.2f} MB')

df = pd.read_csv(RAW_PATH)
print(f'Dataset loaded successfully.')
```

## 2. Shape & Columns

```python
print(f'Rows   : {df.shape[0]:,}')
print(f'Columns: {df.shape[1]}')
print('\nColumn names:')
for i, col in enumerate(df.columns, 1):
    print(f'  {i:2d}. {col}')
```

## 6. Missing Values — First Look

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing': missing, '%': missing_pct})
missing_df[missing_df['Missing'] > 0].sort_values('%', ascending=False)
```

## 7. Unique Values per Column

```python
for col in df.columns:
    print(f'{col:25s} — {df[col].nunique():>6,} unique')
```

## 8. Duplicate Rows

```python
dup_count = df.duplicated().sum()
print(f'Duplicate rows: {dup_count:,} ({dup_count/len(df)*100:.2f}%)')
```

## 9. Key Observations

- **Row count:** 49,782 rows (meets the 5,000-row minimum)
- **Columns:** 17 analytical columns (meets the 8-column minimum)
- **Target column:** `ReturnStatus` — values are `Returned` / `Not Returned`
- **Quality issues observed:**
  - ~4,978 rows (~10%) have missing `CustomerID` — these also lack `ShippingCost` and `WarehouseLocation`
  - Negative `Quantity` rows exist (cancellations/adjustments) with corresponding negative `UnitPrice`
  - `PaymentMethod` has a typo: `paypall` instead of `PayPal`
  - Some `Discount` values exceed 1.0 (data errors)
- **Date range:** 2020-01-01 to 2025-09-05
- **No duplicate rows** found
- **Raw file is untouched** — all cleaning will be done in `02_cleaning.ipynb`
